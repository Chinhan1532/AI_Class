# MTCar

```python
import pandas as pd
url = '/content/mtcars.csv'
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model= linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']), y=data['mpg'])
print('Giá trị Intercept là:',model.intercept_)
print('Giá trị Coeffient là:',model.coef_)

model.score(X=pd.DataFrame(data['wt']), y=data['mpg'])
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
y_pred=model.predict(X=pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2]).T #Tranpose: chuyển vị
model.fit(X, y = data['mpg'])
print('Hệ số Intercept:', model.intercept_)
print('Hệ số Coeffient:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y = data['mpg']))
```

```python
import numpy as np
range = np.arange(1.5, 5.5, 0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'blue', xlim = (0,7))
plt.plot(range, y_pred, color = 'red')
```

# Projit

```python
import pandas as pd
url = '/50_Startups.csv'
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model= linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['R&D Spend']), y=data['Profit'])
print('Giá trị Intercept là:',model.intercept_)
print('Giá trị Coeffient là:',model.coef_)

model.score(X=pd.DataFrame(data['R&D Spend']), y=data['Profit'])
```

```python
data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='black')
y_pred=model.predict(X=pd.DataFrame(data['R&D Spend']))
plt.plot(data['R&D Spend'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'], data['R&D Spend']**2]).T #Tranpose: chuyển vị
model.fit(X, y = data['Profit'])
print('Hệ số Intercept:', model.intercept_)
print('Hệ số Coeffient:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y = data['Profit']))
```

```python
import numpy as np
range = np.arange(1.5, 200000.5, 0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'R&D Spend', y = 'Profit', figsize = (9,9), color = 'blue', xlim = (0,200000))
plt.plot(range, y_pred, color = 'red')
```
